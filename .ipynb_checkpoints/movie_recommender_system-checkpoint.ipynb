```python
### '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

### My target is to convert this list in below format:
# ['Action', 'Adventure', 'Fantasy', 'SciFi']

# genres data is string : 'List'
# to convert it into list, we can use ast.literal_eval, it removes '' this, hence resulting genres
# becomes List
```

```python
# helper function
import ast

def convert(obj):
    L = []
    for dict in ast.literal_eval(obj):
        L.append(dict['name'])

    return L
    
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# apply convert function to all rows of genres 
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def cast_clean(obj):
    L = []
    counter = 0
    for dict in ast.literal_eval(obj):
        L.append(dict['name'])
        counter += 1

        if counter >= 3:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_clean)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
# for crew, extract only director, in crew column his job is director
def fetch_director(obj):
    L = []

    for dict in ast.literal_eval(obj):
        if (dict['job'] == 'Director'):
            L.append(dict['name'])
            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# converting overview into list
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview'][0]
```

```python
movies.head()
```

```python
## remove space from all the words from genres, keywords, cast, crew
## 'Foyez ahmed' -> FoyezAhmed
movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# create a new column tags by concatenating overview, genres, cast, crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# creating a new dataframe with movie_id, title, tags column
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
# tags is now list, lets convert it into string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
# lets convert everything into lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

## Text Vectorization : Bag of Words method

#### Bag of words method:
* combine all the tags , i.e, tags1 + tags2 + ..., this result a large string
* find the topmost 5000 words that used frequently, i.e, w1, w2, ..., w5000
* Now calculate number of time each words from w1, w2,..., w5000 appears in each tags
* this results a 5000 * 5000 matrix where each row is a vector corresponding to a movie
* in the whole process, we will ignore stop words (the words that are used to make the whole sentence, i.e, and, or, in, ...)

* for this task, lets use CountVectorizer Class of scikit-learn

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# top most words
features_name = cv.get_feature_names_out()
print(features_name)
print(len(features_name)) # number of features
```
